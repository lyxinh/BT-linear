```python
import numpy as np 
import matplotlib.pyplot as plt 

x_data = [150.0,155.0,160.0,165.0,170.0] 
y_data = [45.0,50.0,55.0,60.0,65.0]

# chieu cao
# can nang
```

```python
w = 1.0 

def forward(x):
    return x * w

```

```python
#Loss function 
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

w_list = []
mse_list = []
for w in np.arange(0.0, 6.1, 0.1):
    print("w=", w)
    l_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum /5 )
    w_list.append(w)
    mse_list.append(l_sum / 5)
```

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
